```python
from sklearn.model_selection import train_test_split as ttl
from sklearn.neighbors import KNeighborsClassifier as KNN
```

```python
y=df['quality'].values
x=df.drop('quality',axis=1).values
```

```python
x
```

```python
x_train,x_test,y_train,y_test=ttl(x,y,test_size=0.3,stratify=y,random_state=7)
```

```python
y_test
```

```python
model=KNN(n_neighbors=16)
```

```python
model.fit(x_train,y_train)
```

```python
y_predict=model.predict(x_test)
y_predict
```

```python
model.score(x_test,y_test)
```

```python
zeus=[]
for i in range(1,60):
    model=KNN(n_neighbors=i)
    model.fit(x_train,y_train)
    zeus.append(model.score(x_test,y_test))
```

```python
plt.plot(zeus)
```
